```python

def pltCountplot(cats):
    
    fig, axis = plt.subplots(len(cats) // 2 ,2, figsize=(70,40))  

    index = 0
    for i in range(len(cats) // 2):
        for j in range(2):
            
            ax = sns.countplot(x=cats[index], data=data_eda, ax=axis[i][j])
            
            if cats[index] in ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome' , 'y']:
                for item in ax.get_xticklabels():
                    item.set_rotation(40)
            
            if cats[index] in ['age']:
                for item in ax.get_xticklabels():
                    item.set_rotation(90)
                
            for p in ax.patches:
                height = p.get_height()
                ax.text(p.get_x()+p.get_width()/2.,
                        height + 3,
                        '{:1.2f}%'.format(height/len(data_eda)*100),
                        ha="center") 
            index += 1
```

```python
colonnes_str = ['job', 'marital', 'education', 'default', 'housing', 
                'loan', 'contact', 'month', 'poutcome', 'y'
               ]
```

```python
def normalisationStr(lits):
    d = {}
    i = 0
    for x in lits:
        d[x] = i
        i += 1
    return d
```

```python
dic_job = normalisationStr(data_normal['job'].unique())
dic_marital = normalisationStr(data_normal['marital'].unique())
dic_education = normalisationStr(data_normal['education'].unique())

dic_default = normalisationStr(data_normal['default'].unique())
dic_housing = normalisationStr(data_normal['housing'].unique())
dic_loan = normalisationStr(data_normal['loan'].unique())

dic_contact = normalisationStr(data_normal['contact'].unique())
dic_month = normalisationStr(data_normal['month'].unique())
dic_poutcome = normalisationStr(data_normal['poutcome'].unique())

dic_y = normalisationStr(data_normal['y'].unique())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import datasets  # Optional, for sample data generation

# Assuming 'data_normal' is your Pandas DataFrame containing features and target class

# Option 1: Using loc for feature selection (recommended)
X = data_normal[['age', 'job', 'marital']]
Y = data_normal['y']  # Select the last column (target class)


```

```python
# En supposant que Y est votre variable cible
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)


############################# 1. EntraÃ®nement du modÃ¨le : #################################
# CrÃ©er un modÃ¨le de rÃ©gression logistique avec un paramÃ¨tre de rÃ©gularisation
logreg = LogisticRegression(C=1e5)

# EntraÃ®nez le modÃ¨le sur les caractÃ©ristiques (X) et les classes cibles (Y)
logreg.fit(X, Y)
```

```python
# Assuming 'new_age', 'new_job_value', and 'new_marital_value' are your new data points

# Example data point for prediction (using original integer values)
new_data = data_normal[['age', 'job', 'marital']]

# Prediction using the original values
prediction = logreg.predict(new_data)

# 'prediction' will now contain the predicted class label for your new data point
```

```python
########################### 2. PrÃ©paration des donnÃ©es pour la prÃ©diction : #################
# Define the range for the decision boundary plot with a buffer
## DÃ©finir la plage pour le tracÃ© de la frontiÃ¨re de dÃ©cision avec un tampon
x_min, x_max = X.values[:, 0].min() - .5, X.values[:, 0].max() + .5
y_min, y_max = X.values[:, 1].min() - .5, X.values[:, 1].max() + .5
z_min, z_max = X.values[:, 2].min() - .5, X.values[:, 2].max() + .5

# Set the step size for the mesh grid
# DÃ©finir le pas de la grille de maillage
h = .02

# Create a mesh grid for plotting the decision boundary
# CrÃ©er une grille de maillage pour le tracÃ© de la frontiÃ¨re de dÃ©cision
xx, yy, zz = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h), np.arange(z_min, z_max, h))
```

```python
############################# 3. RÃ©alisation de la prÃ©diction :  ####################
# Prediction data (including 'marital')
grid_data = np.c_[xx.ravel(), yy.ravel(), zz.ravel()]
Z = logreg.predict(grid_data)


########################### 4. Visualisation des prÃ©dictions :  #####################
# Reshape the predictions back into the grid dimensions for plotting
Z = Z.reshape(xx.shape)
```

```python
# Create a new plot figure
plt.figure(1, figsize=(6, 6))

# Plot the decision boundary as a colored mesh
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot the original data points with colors based on their actual classes
plt.scatter(X.values[:, 0], X.values[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Paired)

# Set labels for the axes
plt.xlabel('Age')  # Replace with actual feature names if available
plt.ylabel('Job')  # Replace with actual feature names if available

# Adjust plot limits to match the mesh grid
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

# Remove unnecessary ticks for a cleaner visualization
plt.xticks(())
plt.yticks(())

# Display the final plot
plt.show()
```
