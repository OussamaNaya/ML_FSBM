```python
#analyse univariée de la variable cible 
import seaborn as sns
import matplotlib.pyplot as plt

# Histogramme du prix
sns.histplot(df['Prix'], kde=True)
plt.title('Distribution des Prix')
plt.xlabel('Prix')
plt.ylabel('Nombre de Voitures')
plt.show()

# Boîte à moustaches pour voir les valeurs aberrantes
sns.boxplot(x=df['Prix'])
plt.title('Boîte à Moustaches des Prix')
plt.xlabel('Prix')
plt.show()
```

```python
# Convertir la colonne "Prix" en type de données numériques
df['Prix'] = df['Prix'].astype('int64')

# Calcul des quartiles
Q1 = df['Prix'].quantile(0.25)
Q3 = df['Prix'].quantile(0.75)

# Calcul de l'amplitude interquartile (IQR)
IQR = Q3 - Q1

# Définition des limites pour les valeurs aberrantes
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Exclusion des valeurs aberrantes
cleaned_df = df[(df['Prix'] >= lower_bound) & (df['Prix'] <= upper_bound)]

# Affichage des statistiques après exclusion des valeurs aberrantes
print("Nombre de lignes avant exclusion des valeurs aberrantes :", len(df))
print("Nombre de lignes après exclusion des valeurs aberrantes :", len(cleaned_df))

# Vous pouvez utiliser maintenant cleaned_df pour continuer votre analyse
```

```python
# Exclure les colonnes 'colonne1' et 'colonne2' de la DataFrame
df2 = cleaned_df.drop(['Unnamed: 0', 'Lien', 'Ville', 'Airbags', 'Climatisation', 
                      'Système de navigation/GPS', 'Toit ouvrant', 'Sièges cuir', 
                      'CD/MP3/Bluetooth', 'Radar de recul', 'Caméra de recul', 
                      'Vitres électriques', 'ABS', 'ESP', 'Régulateur de vitesse', 
                      'Limiteur de vitesse', 'Ordinateur de bord',"Jantes aluminium"], axis=1)
df2
```

```python
import numpy as np

def kilometrage_numeric(km_str):
    # Vérifie si km_str est une chaîne de caractères
    if isinstance(km_str, str):
        km_range = km_str.split(' - ')
        if len(km_range) == 2:
            return (int(km_range[0].replace(' ', '')) + int(km_range[1].replace(' ', ''))) / 2
        else:
            return np.nan
    else:
        # Si km_str n'est pas une chaîne, on suppose qu'il s'agit déjà d'une valeur numérique ou NaN
        return km_str

# Appliquer la fonction modifiée et remplir les NaN ensuite
df2['Kilométrage'] = df2['Kilométrage'].apply(kilometrage_numeric)

# Il est important de recalculer la moyenne après la conversion car de nouvelles valeurs NaN peuvent avoir été introduites
df2['Kilométrage'] = df2['Kilométrage'].fillna(df2['Kilométrage'].mean())

# Afficher les premières lignes pour vérification
print(df2.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Convertir les variables catégorielles en variables numériques
encoder = LabelEncoder()
df2['Marque'] = encoder.fit_transform(df2['Marque'])
df2['Modèle'] = encoder.fit_transform(df2['Modèle'])
df2['Type de carburant'] = encoder.fit_transform(df2['Type de carburant'])
df2['Boite de vitesses'] = encoder.fit_transform(df2['Boite de vitesses'])
df2['État'] = encoder.fit_transform(df2['État'])
df2['Verrouillage centralisé à distance'] = encoder.fit_transform(df2['Verrouillage centralisé à distance'])
df2
```

```python
from sklearn.model_selection import train_test_split

# Diviser les données en ensembles d'entraînement et de test (par exemple, 80% pour l'entraînement et 20% pour le test)
X_train, X_test, y_train, y_test = train_test_split(df2.drop('Prix', axis=1), df2['Prix'], test_size=0.2, random_state=0)

print("Taille de l'ensemble d'entraînement:", X_train.shape, y_train.shape)
print("Taille de l'ensemble de test:", X_test.shape, y_test.shape)

# X_train : Features pour l'ensemble d'entraînement
# y_train : Labels pour l'ensemble d'entraînement
# X_test  : Features pour l'ensemble de test
# y_test  : Labels pour l'ensemble de test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Initialisation du modèle de régression linéaire
modele_regression = LinearRegression()

# Entraînement du modèle sur les données d'entraînement
modele_regression.fit(X_train, y_train)

# Prédiction sur les données de test
predictions = modele_regression.predict(X_test)

# Calcul de l'erreur quadratique moyenne (MSE) pour évaluer les performances du modèle
mse = mean_squared_error(y_test, predictions)
print("Erreur quadratique moyenne (MSE) :", mse)
```

```python
import matplotlib.pyplot as plt

# Tracer le nuage de points des valeurs réelles par rapport aux valeurs prédites
plt.figure(figsize=(8, 6))
plt.scatter(y_test, predictions, color='blue', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)  # Ligne diagonale pour une prédiction parfaite
plt.xlabel('Valeurs Réelles')
plt.ylabel('Valeurs Prédites')
plt.title('Comparaison des Valeurs Réelles et Prédites')
plt.show()
```
