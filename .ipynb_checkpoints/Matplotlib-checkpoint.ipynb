<h1>Introduction :</h1>
    Matplotlib est une bibliothèque de traçage graphique de bas niveau en python qui sert d'utilitaire de visualisation.
    Matplotlib a été créé par John D. Hunter.
    Matplotlib est open source et nous pouvons l'utiliser librement.
    Matplotlib est principalement écrit en python, quelques segments sont écrits en C, Objective-C et Javascript pour    lacompatibilité avec la plateforme.

```python
import matplotlib

print(matplotlib.__version__)
```

<h3>Exercice : Tracer les points x et y</h3>
    Si nous devons tracer une ligne de (1, 3) à (8, 10), nous devons passer deux tableaux [1, 8] et [3, 10] à la fonction plot.

```python
import matplotlib.pyplot as plt
import numpy as np

axe_x = np.array([1,8])
axe_y = np.array([3,10])

plt.plot(axe_x, axe_y)
plt.show()
```

<h3>Exercice : Traçage sans ligne </h3>
    Pour tracer uniquement les marqueurs, vous pouvez utiliser le paramètre de notation de chaîne de raccourci « o », qui signifie « anneaux ».

```python
import matplotlib.pyplot as plt
import numpy as np

valeurs_x = np.array([1,99])
valeurs_y = np.array([4,17])

plt.plot(valeurs_x, valeurs_y, "o")
plt.show()
```

<h3>Exercice : Points multiples </h3>
     Vous pouvez tracer autant de points que vous le souhaitez, assurez-vous simplement d'avoir le même nombre de points sur les deux axes.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

<h3>Exercice : Points X par défaut </h3>
    Si nous ne précisons pas les points sur l'axe des x, ils obtiendront les valeurs par défaut 0, 1, 2, 3 etc., en fonction de la longueur des points y.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```
